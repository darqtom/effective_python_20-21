## Własności wykresów
Za pomocą specjalnych poleceń można też zmieniać właściwości wykresów, takie jak np.: rodzaj linii, kolor, czy też symbole punktów.

Rodzaje linii:
- \- - ciągła,
- : - punktowa,
- -. - kropka-kreska,
- -- - kreskowana.

Kolory:
- y - yellow,
- m - magenta,
- c - cyan,
- r - red,
- g - green,
- b - blue,
- w - white,
- k - black.

Symbole:
- . - punkt,
- o - kółko,
- x - x,
- \+ - plus,
- \* - gwiazdka,
- s - kwadraty,
- d - romb,
- v - trójkąt w dół,
- ^ - trójkąt w górę,
- < - trójkąt w lewo,
- \> - trójkąt w prawo,
- p - pięciokąt,
- h - sześciokąt.

```python
# własne okienko o rozmiarze 8x6 cali, 80 punktów na cal
plt.figure(figsize=(8, 6), dpi=80)

X = np.linspace(-2*np.pi, 2*np.pi, 256, endpoint=True)
S = np.sin(X)

# ustawiamy kolor, grubość i styl linii
plt.plot(X, S, color="blue", linewidth=2.5, linestyle="-") 

# ustawienie granic osi X
plt.xlim(-8.0, 8.0)

# ustawienie granic osi Y
plt.ylim(-1.0, 1.0)

# podziałka osi X
plt.xticks(np.linspace(-8, 8, 6, endpoint=True))

# podziałka osi Y
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

plt.show()
```

```python
import matplotlib.mlab as mlab

mu = 10  
sigma = 15 
x = mu + sigma * np.random.randn(10000)

num_bins = 30

# alpha - określa przeźroczystość w skali od 0 do 1
n, bins, patches = plt.hist(x, num_bins, normed=True, facecolor='blue', alpha=0.5)

# dorysowanie wykresy gęstości na podstawie histogramu i rozkładu
y = mlab.normpdf(bins, mu, sigma)
plt.plot(bins, y, 'r--', linewidth=2)
plt.ylabel('Prawdopodobienstwo')
plt.title(r'Histogram: $\mu=10$, $\sigma=15$')

plt.show()
```

# Multiwykresy

Podobnie jak w Matlabie możemy tworzyć kilka okien graficznych, umieszczając w nich pojedyncze wykresy. Do tworzenia nowego okno służy funkcja figure. Każde okno jest numerowane od 1, a nie od 0 (inaczej niż w klasycznym indeksowaniu Pythona).

```python
X = np.linspace(-2*np.pi, 2*np.pi, 256, endpoint=True)
S = np.sin(X)
C = np.cos(X)

# pierwsze okno graficzne
plt.figure(1)
plt.plot(X, S)

# drugie okno graficzne
plt.figure(2)
plt.plot(X, C)

plt.show()
```

Niektóre parametry okna graficznego:
- **figsize** - rozmiar okna (szerokość, wysokość) w calach,
- **dpi** - rozdzielczość w punktach na cal,
- **facecolor** - kolor tła,
- **edgecolor** - kolor krawędzi ramki,
- **frameon** - czy rysować ramkę w okół wykresu.

```python
plt.close('all')

# plt.figure(figsize=(8, 6), dpi=80, facecolor='g')
# plt.figure(figsize=(8, 6), dpi=80, facecolor=(0.5, 0.5, 0.5))
plt.figure(edgecolor='r', linewidth=10)

# ax = plt.axes(frameon=False)
# ax.axis('off')


X = np.linspace(-2*np.pi, 2*np.pi, 256, endpoint=True)
S = np.sin(X)

plt.plot(X, S, "b-") 

plt.show()
```

```python
plt.close('all')

def randrange(n, vmin, vmax):
    return (vmax - vmin)*np.random.rand(n) + vmin

ax = plt.subplot(projection='3d')
n = 100
for c, m, zl, zh in [('r', 'o', -50, -25), ('b', '^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zl, zh)
    ax.scatter(xs, ys, zs, c=c, marker=m)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
from matplotlib.animation import FuncAnimation

fig = plt.figure(figsize=(6,6), facecolor='white')
ax = fig.add_axes([0.005,0.005,.99,.99], frameon=True, aspect=1)

# liczba baniek 
n = 50
size_min = 50
size_max = 50*50

# położenie 
P = np.random.uniform(0,1,(n,2))

# Kolory
C = np.ones((n,4)) * (0,0,0,1) 
C[:,3] = np.linspace(0,1,n)

# rozmiar 
S = np.linspace(size_min, size_max, n)

# plot 
scat = ax.scatter(P[:,0], P[:,1], s=S, lw = 0.5,
                  edgecolors = C, facecolors='None')

ax.set_xlim(0,1), ax.set_xticks([])
ax.set_ylim(0,1), ax.set_yticks([])


def update(frame):
    global P, C, S

    C[:,3] = np.maximum(0, C[:,3] - 1.0/n)

    S += (size_max - size_min) / n

    i = frame % 50
    P[i] = np.random.uniform(0,1,2)
    S[i] = size_min
    C[i,3] = 1

    scat.set_edgecolors(C)
    scat.set_sizes(S)
    scat.set_offsets(P)
    return scat,

animation = FuncAnimation(fig, update, interval=10)
# animation.save('banki.gif', writer='imagemagick', fps=30, dpi=72)
plt.show()
```

```python
# trójkąt Sierpińskiego

%matplotlib inline
import random
 
verticles = ( (0,0), (50,75), (100,0) )
start_point = (50,50)
iterations = 10000
 
data_x = []
data_y = []
 
for i in range(iterations):
    k = random.randint(0,2)
    #wez wspolrzedne k-tego wierzcholka
    v_x,v_y = verticles[k]
    #wez wspolrzedne punktu
    p_x, p_y = start_point
    #oblicz nowy punt
    n_x = (v_x + p_x) * 0.5
    n_y = (v_y + p_y) * 0.5
 
    #zapamietaj dane w tablicy 2D
    data_x.append(n_x)
    data_y.append(n_y)
 
    #ustaw punkt startowy jako punkt obliczony
    start_point = (n_x, n_y)
 
plt.plot(data_x, data_y, 'r,')
plt.show()
```
